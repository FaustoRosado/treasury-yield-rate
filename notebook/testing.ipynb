```python
import pandas as pd
import matplotlib.pyplot as plt


# def get_yield_curve():
url = 'https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yield'
yield_curve = pd.read_html(url, parse_dates = True) 
    
```

```python
yield_curve
```

```python
yield_curve = yield_curve[1].dropna(how = 'any')
yield_curve
```

```python
yield_curve_10_2 = yield_curve.iloc[:,[0,6,10]]
yield_curve_10_2.columns = ['date', '2 yr', '10 yr']


#scaled_inputs_all.iloc[shuffled_indices].values 

yield_curve_10_2
```

```python
yield_curve.iloc[:,[6,10]] = yield_curve_10_2[['2 yr', '10 yr']].astype(float)
```

```python
yield_curve_10_2['10yr - 2yr'] = yield_curve_10_2['10 yr'] - yield_curve_10_2['2 yr']
```

```python
yield_curve_10_2
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12, 10))
# plt.plot(yield_curve_10_2['date'], yield_curve_10_2['10 yr'], yield_curve_10_2['2 yr'])
plt.plot(yield_curve_10_2['date'], yield_curve_10_2['10 yr'], "-b", label="10 year")
plt.plot(yield_curve_10_2['date'], yield_curve_10_2['2 yr'], "-r", label="2 year")

plt.title('10 year & 2 year Treasury Rates for April 2021')
plt.xlabel('Date')
plt.xticks(rotation=60, fontsize='medium')
plt.ylabel('Interest Rate')
plt.legend()
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(12, 10))
plt.plot(yield_curve_10_2['date'], yield_curve_10_2['10yr - 2yr'], label='Inline label')
plt.title('10-2 Treasury Yield Spread for April 2021')
plt.xlabel('Date')
plt.xticks(rotation=60, fontsize='medium')
plt.ylabel('10-2 Spread')
plt.legend()
plt.show()
```
